# I.  Principes g√©n√©raux

La m√©thode diviser pour r√©gner consiste, pour r√©soudre un probl√®me de taille N, √† :  

- **Diviser** : partager le probl√®me en sous-probl√®mes ( de taille N-1, parfois de taille N/2, en tous cas de taille plus petite)

- **R√©gner** : r√©soudre ces diff√©rents sous-probl√®mes (g√©n√©ralement r√©cursivement)

- **Combiner** : fusionner les solutions pour obtenir la solution du probl√®me initial

# II. Exemple de recherche d'un √©l√©ment dans une liste non tri√©e

Le probl√®me est de rechercher la pr√©sence d‚Äôun √©l√©ment dans une liste non tri√©e.  

En adoptant le paradigme "diviser pour r√©gner", l‚Äôid√©e pour r√©soudre cette question est de rechercher r√©cursivement l‚Äô√©l√©ment dans la premi√®re moiti√© de la liste et dans la seconde, puis de combiner les r√©sultats via l‚Äôop√©rateur logique or.  

En effet, l‚Äô√©l√©ment recherch√© sera dans la liste s‚Äôil est dans la premi√®re moiti√© ou dans la seconde. La condition d‚Äôarr√™t √† la r√©cursivit√© sera l‚Äôobtention d‚Äôune liste √† un seul √©l√©ment,car il est alors imm√©diat de conclure si l‚Äô√©l√©ment recherch√© appartient √† une telle liste ou non.

Voici donc les trois √©tapes de la r√©solution de ce probl√®me via la m√©thode "diviser pour r√©gner":

‚Ä¢ **Diviser la liste en deux sous-listes en la ‚Äúcoupant‚Äù par la moiti√©.**

‚Ä¢ **Rechercher la pr√©sence de l‚Äô√©l√©ment dans chacune de ces sous-listes. Arr√™ter la r√©cursion lorsque les listes n‚Äôont plus qu‚Äôun seul √©l√©ment.**

‚Ä¢ **Combiner avec l‚Äôop√©rateur logique <font color="blue">or</font> les r√©sultats obtenus.**

Impl√©menter cet algorithme en Python ci-dessous

```python
def recherche(lst, nombre, d, f):
    """
    Pr√©condition : lst est une liste √† priori non tri√©e, nombre est un √©l√©ment cherch√© dans la liste 
    d est le rang du d√©but de la recherche, et f le rang de la fin de la recherche dans lst
    postcondition : la fonction renvoie du bool√©en : 
    True si nombre est dans la liste, et False sinon.
    """
    pass

ma_liste = [95, 28, 36, 52, 85, 56, 34, 59, 17, 26, 16, 25, 69, 98, 4, 85, 81, 48, 11, 57]
assert recherche(ma_liste , 28, 0, len(ma_liste) - 1) == True
assert recherche(ma_liste , 59, 0, len(ma_liste) - 1) == True
assert recherche(ma_liste , 57, 0, len(ma_liste) - 1) == True
assert recherche(ma_liste , 95, 0, len(ma_liste) - 1) == True
assert recherche(ma_liste , 11, 0, len(ma_liste) - 1) == True
assert recherche(ma_liste , 9, 0, len(ma_liste) - 1) == False
```

## 2. Comparaison avec la m√©thode de parcours s√©quentiel

On donne ici une m√©thode "na√Øve" par parcours de liste s√©quentiel.

```python
def recherche_seq(element, tab):
    """
    tab est une liste d'√©l√©ments de types quelconques
    La fonction renvoie True si element est dans tab, et False sinon
    """
    reponse = False
    for elt in tab: # tab est donc parcouru en entier
        if elt == element:
            reponse = True
    return reponse

ma_liste = [95, 28, 36, 52, 85, 56, 34, 59, 17, 26, 16, 25, 69, 98, 4, 85, 81, 48, 11, 57]
assert recherche_seq(28, ma_liste) == True
assert recherche_seq(59, ma_liste) == True
assert recherche_seq(57, ma_liste) == True
assert recherche_seq(95, ma_liste) == True
assert recherche_seq(11, ma_liste) == True
assert recherche_seq(9, ma_liste) == False
```

Nous allons comparer ces deux m√©thodes dans le pire des cas (nous cherchons un √©l√©ment x qui ne se trouve pas dans la liste):

```python
import timeit
import matplotlib.pyplot as plt
from random import shuffle

# diff√©rentes tailles de listes
abscisse = [1000, 50000, 100000, 500000]

ordonnee_dr = [] # liste des temps par diviser pour r√©gner
ordonnee_seq = [] # liste des temps par parcours s√©quentiel

# Cr√©ation des listes non tri√©es de tailles n, et de l'√©l√©ment x √† chercher
for n in abscisse:
    l = [i for i in range(n)] # cr√©ation de la liste de taille n : [0... n-1]
    shuffle(l) # m√©lange de cette liste
    x = n + 10 # x n'appartient pas √† la liste
    d = 0 
    f = n - 1

    # Cr√©ation des listes d'ordonn√©es correspondantes pour chaque graphique
    # Pour diviser pour r√©gner :
    temps=timeit.timeit("recherche(l, x, d, f)", number=1, globals=globals())
    ordonnee_dr.append(temps)
    # Pa parcours s√©quentiel
    temps=timeit.timeit("recherche_seq(x, l)", number=1, globals=globals())
    ordonnee_seq.append(temps)

# Graphique pour des recherches par diviser pour r√©gner en rouge
plt.plot(abscisse, ordonnee_dr, "ro-") # en rouge

# Graphique pour des recherches par parcours s√©quentiel en bleu
plt.plot(abscisse, ordonnee_seq, "bo-") # en bleu

plt.show()
plt.close()
```

**Corrig√© :**  
Ici, le parcours s√©quentiel est plus efficace. Une explication est que les appels de fonction (r√©cursivement) ont un co√ªt.  
Utiliser la m√©thode diviser pour r√©gner dans ce cas-l√† s'apparente √† "√©craser une mouche avec un marteau-pilon."

## 2. Comparaison avec la m√©thode de parcours s√©quentiel

On donne ici une m√©thode "na√Øve" par parcours de liste s√©quentiel.

```python
def max_iter(tab):
    max = tab[0]
    for elt in tab:
        if elt > max :
            max = elt
    return max

assert max_iter([95, 28, 36, 52, 85, 56, 34, 59, 17, 26, 16, 25, 69, 98, 4, 85, 81, 48, 11, 57]) == 98
```

Nous allons comparer ces deux m√©thodes

```python
import timeit
import matplotlib.pyplot as plt
from random import shuffle,randint

# diff√©rentes tailles de listes
abscisse = [100, 200, 300, 400, 500, 600, 800]

ordonnee_dr = [] # liste des temps par diviser pour r√©gner
ordonnee_seq = [] # liste des temps par parcours s√©quentiel

# Cr√©ation des listes non tri√©es de tailles n, et de l'√©l√©ment x √† chercher
for n in abscisse:
    L = [randint(0, n) for i in range(n)] # liste de n entiers al√©atoires compris entre 0 et n

    # Cr√©ation des listes d'ordonn√©es correspondantes pour chaque graphique
    # Pour diviser pour r√©gner :
    temps=timeit.timeit("maximum_recursif_v2(L, 0, None)", number=100, globals=globals())
    ordonnee_dr.append(temps)
    # Par parcours s√©quentiel
    temps=timeit.timeit("max_iter(L)", number=100, globals=globals())
    ordonnee_seq.append(temps)

# Graphique pour des recherches par diviser pour r√©gner en rouge
plt.plot(abscisse, ordonnee_dr, "ro-") # en rouge

# Graphique pour des recherches par r√©cursivit√© en bleu
plt.plot(abscisse, ordonnee_seq, "bo-") # en bleu

plt.show()
plt.close()


```

**Conclusion :**  
Le parcours s√©quentiel est toujours plus efficace . Une explication est que les appels de fonction (r√©cursivement) ont un co√ªt.

## 3. Comparaison des deux m√©thodes

## 4. Complexit√© de l'exponentiation rapide

On peut montrer que la complexit√© (en regardant le nombre de multiplication) de cet algorithme est en $O(\text{log}_2(N))$ (se prononce "grand O" de  $\text{log}_2(N)$) .  

Nous allons voir cela, mais essayons d'approcher la r√©ponse.

**Combien de multplications ferons nous pour √©valuer expo_dr(2,128) ?**

r√©ponse : 7

$2^{128} = (2^{64})^2  = ((2^{32})^2)^2 = (((2^{16})^2)^2)^2 =  (((2^{8})^2)^2)^2)^2 = ((((2^{4})^2)^2)^2)^2)^2 = (((2^{2})^2)^2)^2)^2)^2$

et finalement :
$2^{128} = ((((((2^{1})^2)^2)^2)^2)^2)^2)^2$

Chaque √©l√©vation au carr√© co√ªte une multiplication. On √©l√®ve 7 fois au carr√©, et donc on √©l√®ve ici √† la puissance $2^7$

Ceci parce qu'on effectue des divisions successives par 2, d'une part, et que 
$2^7=128$ 

Or, $2^7=128$ est equivalent √† $\text{log}_2(128)=7$. Il faudra donc faire 7 multiplications pour calculer $x^{128}$ .
